```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
make_blobs
```

```python
# Create Data
# higher cluster = the point more clust together
data = make_blobs(n_samples=1200, n_features=2, centers=4, cluster_std=11.8,random_state=101)
```

```python
data # 2d
```

```python
print (data[0].shape)
print (data[1].shape)
```

```python
data[0] # 2d
```

```python
data[1] # list
```

## Visualize Data

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
# the green color is pretty distinct from the others. the others clust together
```

```python
x=data[0]
y=data[1]
```

## Creating the Clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(x)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(x[:,0],x[:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(x[:,0],x[:,1],c=y,cmap='rainbow')
```

```python
from sklearn.metrics import classification_report as cr, confusion_matrix as cm
```

```python
print(cm(data[1], kmeans.labels_))
print(cr(data[1], kmeans.labels_))
```
