```python
from sklearn.model_selection import train_test_split
X = hp[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
               'Avg. Area Number of Bedrooms', 'Area Population']]
y=hp['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lrm=LinearRegression()
```

```python
lrm.fit(X_train, y_train)
```

```python
prediction=lrm.predict(X_test)
```

```python
plt.scatter(prediction, y_test)
plt.legend()
```

```python
lrm.intercept_
```

```python
hp_coe=pd.DataFrame(lrm.coef_, X.columns, columns=['Coefficients'])
hp_coe
```

```python
# Is it a good model?
```

```python
sns.jointplot(hp['Avg. Area Income'],y_test-prediction)
sns.jointplot(hp['Avg. Area House Age'],y_test-prediction)
sns.jointplot(hp['Avg. Area Number of Rooms'],y_test-prediction)
```

```python
sns.jointplot(hp['Avg. Area Number of Bedrooms'],y_test-prediction)
sns.jointplot(hp['Area Population'],y_test-prediction)
```

```python
sns.distplot(y_test-prediction, bins=100)
```

```python
# residual shows no correclation with 
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
print('QSE:',lrm.score(X,y) )
```
